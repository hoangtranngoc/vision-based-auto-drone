```python
import cv2
import numpy as np
import scipy.misc
import os
# from os import listdir, walk
# from os.path import isfile, join, splitext
import random
import imageio
```

```python
def create_normal_line_and_line_map(f, path, output_path):
    file, ext = os.path.splitext(f)
    if (ext != ".png"):
        return
    
    img = cv2.imread(os.path.join(path,f))
    img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    # Gen lower mask (0-5) and upper mask (175-180) of RED
    mask1 = cv2.inRange(img_hsv, np.array([0,50,25]), np.array([12,255,255]))
    mask2 = cv2.inRange(img_hsv, np.array([167,50,50]), np.array([179,255,255]))

    mask = mask1 + mask2
    output_hsv = img_hsv.copy()
    output_hsv[np.where(mask==0)] = 0

    img_rgb = cv2.cvtColor(output_hsv, cv2.COLOR_HSV2RGB)
    imageio.imwrite(os.path.join(output_path,file+'_linemap.png'), img_rgb)
    

    ## Merge the mask and crop the red regions
    mask = cv2.bitwise_or(mask1, mask2 )
    croped = cv2.bitwise_and(img_hsv, img_hsv, mask=mask)
    #print (croped.shape)



    r, c, _ = img.shape
    for i in range (0, r):
        for j in range (0, c):
            pixel = croped[i,j,:]
            if (np.array_equal(pixel, [0,0,0]) == False):           
                #print(i,j,pixel)
                if (img_hsv[i,j,0] < 13 and img_hsv[i,j,1] < 200 and img_hsv[i,j,2] < 130):  

                    img_hsv[i,j,1] = random.randint(25,29)*2.55
                    img_hsv[i,j,2] = random.randint(21,25)*2.55               
                else:
                    img_hsv[i,j,1] = random.randint(32, 37)*2.55
                    img_hsv[i,j,2] = random.randint(32, 37)*2.55

                img_hsv[i,j,0] = random.randint(16,22)

    img_rgb = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2RGB)

    imageio.imwrite(os.path.join(output_path,file+'_normal.png'), img_rgb)    
    
```

```python
directory = "E:\\UAV_drone\\data\\training\\lateral_and_view"
sub_dirs = [x[0] for x in os.walk(directory)][1:] #doesn't count the root directory
print(sub_dirs)
for input_path in sub_dirs:

    output_path = os.path.join(input_path, "output")
    if not os.path.exists(output_path):
        os.makedirs(output_path)
    
    onlyfiles = [f for f in os.listdir(input_path) if os.path.isfile(os.path.join(input_path, f))]

    for f in onlyfiles:
        #print(f)
        create_normal_line_and_line_map(f, input_path, output_path)
```

```python
def get_grayscale(filename, max_val_pct=0.1):
    """
    Handle any infinite values, set those equal to largest value + largest value * max_val_pct
    DocTest .pfm from http://vision.middlebury.edu/stereo/data/scenes2014/datasets/Adirondack-perfect/
    :param filename: str
    :param max_val_pct: float
    :return: numpy.array
    >>> grayscale = get_grayscale('images/Adirondack-perfect/disp1.pfm')
    >>> grayscale = cv2.resize(grayscale, dsize=(0, 0), fx=0.2, fy=0.2)
    >>> cv2.imshow('image', cv2.applyColorMap(grayscale, cv2.COLORMAP_JET))
    >>> cv2.waitKey(0)
    """
    data, _ = load_pfm(open(filename, 'rb'))
    data = np.where(data == np.inf, -1, data)
    max_val = np.max(data)
    max_val += max_val * max_val_pct
    data = np.where(data == -1, max_val, data)
    return cv2.normalize(data, dst=None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX).astype(np.uint8)
```
